```python
#Gradiente
def gradF(w):
    y = np.array([
        3*(w[0]**2)*(w[1]**2) - 6*(w[0]**2)*w[1] - 3*(w[1]**2) + 6*w[1],
        2*w[1]*(w[0]**3) - 2*(w[0]**3) - 6*w[0]*w[1] + 6*w[0]
    ])
    return y
```

```python
#Hessiana
def Hessi(w):
    y = np.array([
        [6*w[0]*(w[1]**2) - 12*w[0]*w[1], 6*(w[0]**2)*w[1] - 6*(w[0]**2) - 6*w[1] + 6],
        [6*(w[0]**2)*w[1] - 6*(w[0]**2) - 6*w[1] + 6, 2*(w[0]**3) - 6*w[0]]
    ])
    return y
```
