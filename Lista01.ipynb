```python
#Hessiana
def Hessi(w):
    y = np.array([
        [6*w[0]*(w[1]**2) - 12*w[0]*w[1], 6*(w[0]**2)*w[1] - 6*(w[0]**2) - 6*w[1] + 6],
        [6*(w[0]**2)*w[1] - 6*(w[0]**2) - 6*w[1] + 6, 2*(w[0]**3) - 6*w[0]]
    ])
    return y
```

```python
def F(w1, w2):
    y = ((w1**2 + w2**2 - 1)*w1*w2)**2
    return y
```

```python
def e (w1, w2):
    y = ((w1**2 + w2**2 - 1)*w1*w2)
    return y
```

```python
def dedw1 (w1, w2):
    y = (3*w1**2)*w2 + w2**3 - w2
    return y
```

```python
def dedw2 (w1, w2):
    y = (3*w2**2)*w1 + w1**3 - w1
    return y
```

```python
def d2edw12 (w1, w2):
    y = 6*w1*w2
    return y
```

```python
def d2edw1w2 (w1, w2):
    y = 3*w1**2 + 3*w2**2 - 1
    return y
```

```python
def h (w1, w2):
    y = 2*(dedw1(w1, w2))*(dedw2(w1,w2))
    return y
```

```python
def dFdw1 (w1, w2):
    y = 2*eps*dedw1(w1, w2)
    return y
```

```python
def dFdw2 (w1, w2):
    y = 2*eps*dedw2(w1, w2)
    return y
```

```python
def d2Fdw1w2 (w1, w2):
    y = 2*(dedw1(w1,w2)*dedw2(w1,w2)+eps*d2edw1w2(w1,w2))
    return y
```
